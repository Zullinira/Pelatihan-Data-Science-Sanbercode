##### cara explicit 

(di rekomendasikan ketika kita ingin membuat visualisasi data yang lebih kompleks.)

```python
#membuatobjek figure
fig=plt.figure()

#membuat objek axes
ax=fig.add_subplot()

#membuat data
data_x = [1,2,3,4,5]
data_y = [10,20,25,30,15]

#menmabahkan dalam obkel axes
ax.plot(data_x, data_y)

#menmpilkan data
plt.show()
```

#### cara implisit
cara lain yang lebih clean

```python
plt.plot(data_x, data_y)

plt.show()
```

#### dengan subplots()

```python
#membuat figure dan axis
fig, ax=plt.subplots()

 #data
data_x = [1, 2, 3, 4, 5]
data_y = [10, 20, 25, 30, 15]

#menambahkan dalamobjek axes
ax.plot(data_x,data_y)

#tampilkan
plt.show()
```

##### memplot lebih dari satu data kedalam suatu axis

```python
fig,ax =plt.subplots()
data_x1 = [1, 2, 3, 4]
data_y1 = [10, 20, 25, 30]
data_x2 = [1, 2, 3, 4, 5]
data_y2 = [5, 15, 20, 25, 30]

ax.plot(data_x1,data_y1)
ax.plot(data_x2, data_y2)
plt.show()
```

### Menyimpan dan menampilkan gambar hasil plotting

```python
#membuat figuer and axis

fig,ax =plt.subplots()
data_x1 = [1, 2, 3, 4, 5]
data_y1 = [10, 20, 25, 30, 35]
data_x2 = [1, 2, 3, 4, 5]
data_y2 = [5, 15, 20, 25, 30]

ax.plot(data_x1, data_y1)
ax.plot(data_x2, data_y2)

plt.show()
```

```python
import numpy as np
import random
import  matplotlib.pyplot as plt

fig,ax=plt.subplots()

data_x1 = [1, 2, 3, 4, 5]
data_y1 = [10, 20, 25, 30, 35]
data_x2 = [1, 2, 3, 4, 5]
data_y2 = [5, 15, 20, 25, 30]

ax.plot(data_x1,data_y1)
ax.plot(data_x2, data_y2)

plt.style.use('seaborn')

plt.show()
```

```python
#cara implisit
plt.figure(figsize=(7,4))
plt.plot(data_x1, data_y1)
plt.plot(data_x2, data_y2)

plt.style.use('default')
plt.show()
```
