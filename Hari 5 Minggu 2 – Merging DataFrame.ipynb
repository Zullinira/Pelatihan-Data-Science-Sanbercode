```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                            'B': ['B0', 'B1', 'B2', 'B3'],
                            'C': ['C0', 'C1', 'C2', 'C3'],
                            'D': ['D0', 'D1', 'D2', 'D3']},
                           index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                            'B': ['B4', 'B5', 'B6', 'B7'],
                            'C': ['C4', 'C5', 'C6', 'C7'],
                            'D': ['D4', 'D5', 'D6', 'D7']},
                           index=[4, 5, 6, 7])

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                            'B': ['B8', 'B9', 'B10', 'B11'],
                            'C': ['C8', 'C9', 'C10', 'C11'],
                            'D': ['D8', 'D9', 'D10', 'D11']},
                           index=[8, 9, 10, 11])
```

```python
display(df1)
display(df2)
display(df3)
```

menggabungkan baris data ini, kita bisa menggunakan suatu method concat(). pertama kita buat terlebih dahulu list dari data yang akan kita gabung, kemudian masukan list tersebut sebagai argument kedalam method concat(). 

```python
list_df=[df1,df2,df3]
```

```python
list_df
```

```python
pd.concat(list_df)
```

 menggabungkan data tersebut berdasarkan column

```python
pd.concat(list_df, axis=1)
```

### Logika-Logika Penggabungan Dengan Metode merge

### Full Outer Join

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                              'A': ['A0', 'A1', 'A2', 'A3'],
                              'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K4', 'K5'],
                                 'C': ['C0', 'C1', 'C2', 'C3'],
                                 'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
print('data pertama')
print('================================')
left
```

```python
print('data kedua')
print('================================')
right
```

```python
# gabungkan data dengan logika outer join/full join
pd.merge(left, right, on='key', how='outer')
```

### Inner Join

```python
pd.merge(left,right, on='key', how='inner')
```

### Left Join

```python
pd.merge(left,right,on='key', how='left')
```

### Right Join

```python
pd.merge(left, right, on='key', how='right')
```
